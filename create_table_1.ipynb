```python
#hb = homebased
# hb = raw['HOMEBASED']
raw_hb = raw[raw['HOMEBASED'] != b'0']
```

```python
raw_hb.shape
```

```python
# hb.isna().sum()
```

```python
hb_1 = raw_hb[raw_hb['HOMEBASED'] == b'1']
```

```python
hb_yes = hb_1.shape[0]
hb_yes
```

```python
hb_2 = raw_hb[raw_hb['HOMEBASED'] == b'2']
```

```python
hb_no = hb_2.shape[0]
```

```python
hb_data = [hb_yes, hb_no]
```

```python
hb_mean = np.mean(hb_data)
hb_std = np.std(hb_data)
hb_percentile_5th = np.percentile(hb_data, 5)
hb_percentile_95th = np.percentile(hb_data, 95)
```

```python
print(f"Mean: {hb_mean}")
print(f"Standard Deviation: {hb_std}")
print(f"5th Percentile: {hb_percentile_5th}")
print(f"95th Percentile: {hb_percentile_95th}")
```
